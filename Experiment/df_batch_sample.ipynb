```python
# Set up simple matrix multiplication
import numpy as np
import matplotlib.pyplot as plt

# Set up the matrices
A = np.array([[1, 2, 3, 5], [2, 4, 8, 12], [3, 6, 7, 13]])
C = np.array([[1, 2], [2, 3], [3, 4]])
M = np.array([[1, 0, 1, 0], [0, 1, 0, 0], [0, 1, 0, 1]])
```

```python
# Optimize for X without using mask M
X = np.ones((2, 4))
mse = []
for i in range(1000):
    X = X - 0.01 * C.T @ (C @ X - A)
    mse.append(np.mean((C @ X - A) ** 2))

# Plot the MSE
plt.plot(mse)
plt.xlabel('Iteration')
plt.ylabel('MSE')
plt.show()
```

```python
# Optimize for X using mask M
X = np.ones((2, 4))
mse = []
for i in range(1000):
    X = X - 0.01 * C.T @ (M * (C @ X) - M * A)
    mse.append(np.mean((M * (C @ X - A)) ** 2))

# Plot the MSE
plt.plot(mse)
plt.xlabel('Iteration')
plt.ylabel('MSE')
plt.show()
```
